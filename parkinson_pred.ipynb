```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from imblearn.combine import SMOTETomek
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split

#feature seperation
X = df_parkinson.drop('status', axis=1)
y = df_parkinson['status']

# splitting the dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=10,stratify=y)

#scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

#sampling
from imblearn.combine import SMOTETomek
smt = SMOTETomek(random_state=10)
X_train_bal, y_train_bal = smt.fit_resample(X_train_scaled, y_train)

model = LogisticRegression(max_iter=1000,)
model.fit(X_train_bal, y_train_bal)
y_pred = model.predict(X_test_scaled)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
train_acc = model.score(X_train_scaled, y_train)
test_acc = model.score(X_test_scaled, y_test)
print(train_acc, test_acc)
```

```python
%%writefile parkinsons_app.py
import streamlit as st
import pandas as pd
import numpy as np
import pickle
from sklearn.preprocessing import StandardScaler

# Load the saved model and components
@st.cache_resource
def load_components():
    with open("model_parkinson.pkl", "rb") as f:
        model = pickle.load(f)
    with open("scaler_parkinson.pkl", "rb") as f:
        scaler = pickle.load(f)
    with open("feature_names_parkinson.pkl", "rb") as f:
        feature_names = pickle.load(f)
    return model, scaler, feature_names

model, scaler, feature_names = load_components()

# Streamlit app
st.title("Parkinson's Disease Prediction")
st.write("Upload a CSV file containing patient data to predict Parkinson's disease status")

# File upload
uploaded_file = st.file_uploader("Choose a CSV file", type="csv")

if uploaded_file is not None:
    try:
        # Read the uploaded file
        input_df = pd.read_csv(uploaded_file)
        
        # Display the uploaded data
        st.subheader("Uploaded Data Preview")
        st.write(input_df.head())
        
        # Check if required columns are present
        missing_cols = set(feature_names) - set(input_df.columns)
        if missing_cols:
            st.error(f"Error: The uploaded file is missing the following columns: {', '.join(missing_cols)}")
        else:
            # Prepare the data (drop 'name' column if exists, but keep others)
            if 'name' in input_df.columns:
                input_df = input_df.drop(columns=['name'])
            
            # Ensure columns are in correct order
            input_data = input_df[feature_names]
            
            # Scale the data
            scaled_data = scaler.transform(input_data)
            
            # Make predictions
            predictions = model.predict(scaled_data)
            prediction_probs = model.predict_proba(scaled_data)
            
            # Add predictions to dataframe
            results_df = input_df.copy()
            results_df['Prediction'] = predictions
            results_df['Prediction'] = results_df['Prediction'].map({0: 'Negative', 1: 'Positive'})
            results_df['Probability (Positive)'] = prediction_probs[:, 1]
            
            # Display results
            st.subheader("Prediction Results")
            st.write(results_df)
            
            # Download results
            csv = results_df.to_csv(index=False).encode('utf-8')
            st.download_button(
                label="Download Predictions as CSV",
                data=csv,
                file_name='parkinsons_predictions.csv',
                mime='text/csv'
            )
            
            # Show summary statistics
            st.subheader("Prediction Summary")
            positive_count = sum(predictions)
            negative_count = len(predictions) - positive_count
            st.write(f"Positive Predictions (Parkinson's): {positive_count}")
            st.write(f"Negative Predictions (No Parkinson's): {negative_count}")
            
    except Exception as e:
        st.error(f"An error occurred: {str(e)}")

# Add some information about the model
st.sidebar.title("About")
st.sidebar.info(
    """
    This application predicts Parkinson's disease status using a pre-trained Logistic Regression model.
    
    **Input Requirements:**
    - CSV file with the same features used during model training
    - Should contain all the required features (excluding 'name' column)
    
    **Output:**
    - Prediction (Positive/Negative)
    - Probability score for positive prediction
    """
)
```
